```python
import gensim
from gensim.models import Word2Vec
```

```python
df_playstyle = df_clean.Playstyle.apply(gensim.utils.simple_preprocess)
```

```python
df_playstyle
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2,
    workers=4,
)
```

```python
model.build_vocab(df_playstyle, progress_per=1000)
```

```python
model.train(df_playstyle, total_examples=model.corpus_count, epochs=model.epochs)
```
